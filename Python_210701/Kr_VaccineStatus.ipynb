```python
import requests 
from bs4 import BeautifulSoup
```

```python
url = "https://ncv.kdca.go.kr/mainStatus.es?mid=a11702000000"
htmlText = requests.get(url).text
bsoup = BeautifulSoup(htmlText, "html.parser")
btrs = bsoup.find_all("tr")
# btdcols = btrs[7].find_all("td",{"class": "ta_r"})
btds = btrs[7:]

vaccine_data = []

for bc in btds :
    dic = {}
    tdc = bc.find_all("th")
    dic['city'] = tdc[0].text    
    vaccine_data.append(dic)
```

```python
for bc in btds :
    dic = {}
    tdc = bc.find_all("th")
    dic['city'] = tdc[0].text    
    for bd in btds :
        tds = bd.find_all("td")
        dic['first_vaccine'] = tds[1].text
        dic['second_vaccine'] = tds[3].text
        vaccine_data.append(dic)
```
