```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def celsius_to_fahrenheit(x):
  return x* 1.8 + 32
  #정의(함수) 꼭 확인할 필요 있음! 
```

```python
data_C = np.array(range(0, 100),)
data_F = celsius_to_fahrenheit(data_C)
print(data_C, '\n')
print(data_F)
```

```python
inp = int(input("섭씨 온도를 입력하세요.\n"))
print("화씨 온도로", celsius_to_fahrenheit(inp), "입니다.")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(1,input_dim=1))
  
model.compile(loss='mse', optimizer='rmsprop', metrics=['mae'])
print(model.summary())
```

```python
print(model.predict([0.01]))
```

```python
model.save("/content/before_learning.h5")
```

```python
scaled_data_C = data_C / 100
scaled_data_F = data_F / 100
```

```python
print(scaled_data_C)
print(scaled_data_F)
```

```python
fit_hist = model.fit(scaled_data_C, scaled_data_F, epochs = 300, verbose=1)
```

```python
plt.plot(fit_hist.history['loss'])
plt.show()
```

```python
print(model.predict([1])) #학습 많이 하면 overfitting 생김, 즉 적절히!
```

```python
model.save("/content/after_learning.h5")
```

```python
noise = np.array(np.random.normal(0, 0.1, 100))
print(noise)
```

```python
noised_scaled_data_F = scaled_data_F + noise
```

```python
plt.scatter(x=scaled_data_C, y=scaled_data_F)
plt.show()
```

```python
plt.scatter(x=scaled_data_C, y=noised_scaled_data_F)
plt.show()
```

```python
model2 = Sequential()
model2.add(Dense(1,input_dim=1))
  
model2.compile(loss='mse', optimizer='rmsprop', metrics=['mae'])
print(model2.summary())
```

```python
fit_hist2 = model2.fit(scaled_data_C, noised_scaled_data_F, epochs = 300, verbose=1)
```

```python
plt.plot(fit_hist2.history['loss'])
plt.show()
```

```python
print(model2.predict([0.01]))
```
